```python
!pip install missingno
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
import os
import glob

pd.set_option('display.max_columns', None)
print("All libraries loaded!")
```

```python
data_path = "../data/raw/Delhi-20260327T130311Z-1-001/Delhi"

all_files = glob.glob(os.path.join(data_path, "*.csv"))
print(f"Found {len(all_files)} station files:")
for f in all_files:
    print(" ", os.path.basename(f))
```

```python
sample = pd.read_csv(all_files[0])
print("Shape:", sample.shape)
print("\nColumns:", sample.columns.tolist())
print("\nFirst 3 rows:")
sample.head(3)
```

```python
dfs = []

for f in all_files:
    try:
        df = pd.read_csv(f)
        
        # Extract info from filename
        basename = os.path.basename(f)
        parts = basename.replace('.csv', '').split('_')
        
        # Extract year
        year = None
        for part in parts:
            if part.isdigit() and len(part) == 4:
                year = int(part)
                break
        
        # Extract station name
        try:
            site_idx = parts.index('site')
            station_parts = []
            for part in parts[site_idx+2:]:
                if part in ['Delhi', 'CPCB', 'DPCC']:
                    break
                station_parts.append(part)
            station = '_'.join(station_parts)
        except:
            station = basename

        df['Station'] = station
        df['Year'] = year
        df['Source_file'] = basename
        dfs.append(df)
        
    except Exception as e:
        print(f"ERROR loading {basename}: {e}")

print(f"\nLoaded {len(dfs)} files successfully")
combined = pd.concat(dfs, ignore_index=True)
print(f"Total rows: {combined.shape[0]:,}")
print(f"Total columns: {combined.shape[1]}")
print(f"\nYears available: {sorted(combined['Year'].unique().tolist())}")
print(f"Stations available: {combined['Station'].nunique()} unique stations")
print(combined['Station'].unique().tolist())
```

```python
col_rename = {
    'Timestamp': 'Timestamp',
    'PM2.5 (µg/m³)': 'PM25',
    'PM10 (µg/m³)': 'PM10',
    'NO (µg/m³)': 'NO',
    'NO2 (µg/m³)': 'NO2',
    'NOx (ppb)': 'NOx',
    'NH3 (µg/m³)': 'NH3',
    'SO2 (µg/m³)': 'SO2',
    'CO (mg/m³)': 'CO',
    'Ozone (µg/m³)': 'Ozone',
    'Benzene (µg/m³)': 'Benzene',
    'Toluene (µg/m³)': 'Toluene',
    'Xylene (µg/m³)': 'Xylene',
    'O Xylene (µg/m³)': 'O_Xylene',
    'Eth-Benzene (µg/m³)': 'Eth_Benzene',
    'MP-Xylene (µg/m³)': 'MP_Xylene',
    'AT (°C)': 'AT',
    'RH (%)': 'RH',
    'WS (m/s)': 'WS',
    'WD (deg)': 'WD',
    'RF (mm)': 'RF',
    'TOT-RF (mm)': 'TOT_RF',
    'SR (W/mt2)': 'SR',
    'BP (mmHg)': 'BP',
    'VWS (m/s)': 'VWS'
}

combined.rename(columns=col_rename, inplace=True)

# Convert Timestamp to datetime
combined['Timestamp'] = pd.to_datetime(combined['Timestamp'], errors='coerce')

# Drop rows where Timestamp failed to parse
bad_timestamps = combined['Timestamp'].isna().sum()
print(f"Rows with bad timestamps dropped: {bad_timestamps}")
combined.dropna(subset=['Timestamp'], inplace=True)

# Sort
combined.sort_values(['Station', 'Timestamp'], inplace=True)
combined.reset_index(drop=True, inplace=True)

print("\nDone! Final shape:", combined.shape)
print(f"Date range: {combined['Timestamp'].min()} → {combined['Timestamp'].max()}")
```

```python
print("=== MISSING VALUES PER COLUMN (%) ===\n")
missing = (combined.isnull().sum() / len(combined) * 100).round(2)
print(missing)

print("\n=== ROWS PER YEAR ===\n")
print(combined.groupby('Year').size())

print("\n=== ROWS PER STATION ===\n")
print(combined.groupby('Station').size().sort_values(ascending=False))
```

```python
# Save full dataset
out_path = '../data/processed/delhi_all_years_all_stations.csv'
combined.to_csv(out_path, index=False)

size_mb = os.path.getsize(out_path) / (1024*1024)
print(f"Saved to: {out_path}")
print(f"File size: {size_mb:.1f} MB")
print(f"Total rows: {combined.shape[0]:,}")
print(f"Total stations: {combined['Station'].nunique()}")
print(f"Years: {sorted(combined['Year'].unique().tolist())}")
```
